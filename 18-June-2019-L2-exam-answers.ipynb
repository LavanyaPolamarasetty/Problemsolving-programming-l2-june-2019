```python
# write a program to determine whether a  number N is equal to the sum of its prpper positive divisor.
def isprime(f):
    c=0
    sum=0
    #print(f)
    for j in range(1,f):
        #print(j)
        if f%j==0:
            c+=1
            sum+=j
    return sum
  
def perfectnum(n):
    if isprime(n)==n:
        print("YES")
    else:
        print("NO")

t=int(input())
for i in range(t):
    n=int(input())
    perfectnum(n)
```
