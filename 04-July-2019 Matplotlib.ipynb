```python
def getcolumnindex(applestoredf,columnkey):
    columnindex=0
    for i in range(len(applestoredf.columns)):
        if applestoredf.columns[i]==columnkey:
            columnindex=i
    return columnindex
getcolumnindex(applestoredf,"prime_genre")
```

```python
#  Voltage vs  time
import numpy as np
t = [i/10 for i in range (1,21)]
v=[1+np.sin(2*np.pi*i) for i in t]
plt.title("Voltage vs Time")
plt.xlabel("Time")
plt.ylabel("Voltage")
plt.plot(t,v)
plt.savefig("images_graphs/Voltage-Time.png")
plt.show()
```

```python
# Plot Income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv('Files/income.csv')
years=df.columns[2:]
#years=list(range(2005,2014))
years
incomeArkansas = df.values[3,2:]
plt.title("Income of Akansas")
plt.xlabel("Year")
plt.ylabel('Income')
plt.plot(years,incomeArkansas)
plt.show()
```
